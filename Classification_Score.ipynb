```python
def scoring(t, predicted, actual):
  print(f'{t} Confusion Matrix: \n {confusion_matrix(actual,predicted)}')
  print(f'{t} Accuracy: {accuracy_score(actual,predicted)}')
  print(f'{t} Precision:  {precision_score(actual,predicted)}')
  print(f'{t} Recall: {recall_score(actual,predicted)}')
def score(model, X_train, X_test, Y_train, Y_test):
  test_pred = model.predict(X_test)
  train_pred = model.predict(X_train)
  scoring("Test", test_pred, Y_test)
  scoring("Train", train_pred, Y_train)
```
